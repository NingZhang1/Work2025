```python
import pyscf_util
from pyscf_util.Analyzer.iCIPT2.analyzer import *
from pyscf_util.misc.picture import *
```

```python
import os

Res = {
    
}

# Loop through all files in the ../DZ/ directory
for filename in os.listdir('../DZ/'):
    if filename.endswith('.out'):
        filepath = os.path.join('../DZ/', filename)
        print(f"Analyzing file: {filepath}")
        # Perform analysis on the output file
        res = extract_icipt2_data_from_file2(filepath)  # Assuming analyze_output is a function from pyscf_util
        # print(res)
        
        ept = [x.ept for x in res]
        etot = [x.etot for x in res]
        print(ept)
        print(etot)
        print("------------------------------")
        print(extra_icipt2_data(res, 3))
        reg3_res = extra_icipt2_data(res, 3)
        print("------------------------------")
        print(extra_icipt2_data(res, 4))
        reg4_res = extra_icipt2_data(res, 4)
        print("------------------------------") 
        print(extra_icipt2_data(res, 5))
        reg5_res = extra_icipt2_data(res, 5)
        print("------------------------------")
        
        filename = filename.replace('.out', '')
        task_name, type_name, basis = filename.split('_')
        
        if task_name not in Res:
            Res[task_name] = {}
        if type_name not in Res[task_name]:
            Res[task_name][type_name] = {}
        if basis not in Res[task_name][type_name]:
            Res[task_name][type_name][basis] = {}
        
        Res[task_name][type_name][basis]['ept'] = ept
        Res[task_name][type_name][basis]['etot'] = etot
        Res[task_name][type_name][basis]['reg3'] = {
            'res' : reg3_res.weighted_linear_extra,
            'err' : reg3_res.weighted_linear_error,
        }
        
        Res[task_name][type_name][basis]['reg4'] = {    
            'res' : reg4_res.weighted_linear_extra,
            'err' : reg4_res.weighted_linear_error,
        }
        
        Res[task_name][type_name][basis]['reg5'] = {
            'res' : reg5_res.weighted_linear_extra,
            'err' : reg5_res.weighted_linear_error,
        }
        
        ### plot ###
        
        # draw_extra_pic([ept], [etot], title=filename, line_prop=[
        #     {
        #         'marker': 'o',
        #         'markersize': 12,
        #         'linewidth': 3
        #     }],
        #     legend= ['iCIPT2_Data']
        # )
```
